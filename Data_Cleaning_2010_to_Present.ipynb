```python
import pandas as pd
```

```python
df1=pd.read_csv('Crime_Data_from_2010_to_2019_20241118.csv')
df2=pd.read_csv('Crime_Data_from_2020_to_Present_20241118.csv')
print(df1.head()) 
```

```python
common_cols = df1.columns.intersection(df2.columns)  
unique_to_df1 = df1.columns.difference(df2.columns)  
unique_to_df2 = df2.columns.difference(df1.columns)  

print("Common columns:", common_cols.tolist())  
print("\nColumns unique to df1:", unique_to_df1.tolist())
print("Columns unique to df2:", unique_to_df2.tolist())
```

```python
df1.rename(columns={'AREA ' : 'AREA' }, inplace=True)
```

```python
df= pd.concat([df1,df2], ignore_index=True)
df.info()
```

```python
df.columns=df.columns.str.strip().str.replace(" ","_").str.lower()
df.info()
```

```python
df.drop(columns=['crm_cd_1','crm_cd_2','crm_cd_3','crm_cd_4','part_1-2','cross_street','area','mocodes','crm_cd','premis_cd','weapon_used_cd','status'], inplace=True)
```

```python
df.rename(columns={'lat':'latitude','lon':'longitude','crm_cd':'crime_code','rpt_dist_no':'district_no','dr_no':'doc_no','vict_descent':'vict_race'})
```

```python
df.fillna("null")
```

```python
df = df.drop_duplicates(subset=['dr_no'])
```

```python

df['date_rptd'] = pd.to_datetime(df['date_rptd'], format='%Y-%m-%d', errors='coerce')
df['date_occ'] = pd.to_datetime(df['date_occ'], format='%Y-%m-%d', errors='coerce')
```

```python
df.columns = df.columns.str.strip().str.lower()
if 'time_occ' in df.columns:
    df['time_occ'] = df['time_occ'].astype(str)  
    df['time_occ'] = df['time_occ'].str.zfill(4) 
    df['time_occ'] = pd.to_datetime(df['time_occ'], format='%H%M').dt.time 
else:
    print("Column 'time_occ' not found in the DataFrame")
```

```python
race_list = {
    'A': 'Other Asian',
    'B': 'Black',
    'C': 'Chinese',
    'D': 'Cambodian',
    'F': 'Filipino',
    'G': 'Guamanian',
    'H': 'Hispanic/Latin/Mexican',
    'I': 'American Indian/Alaskan Native',
    'J': 'Japanese',
    'K': 'Korean',
    'L': 'Laotian',
    'O': 'Other',
    'P': 'Pacific Islander',
    'S': 'Samoan',
    'U': 'Hawaiian',
    'V': 'Vietnamese',
    'W': 'White',
    'X': 'Unknown',
    'Z': 'Asian Indian'
}

# Map ethnicity and sex directly
df['vict_descent'] = df['vict_descent'].map(race_list).fillna('Unknown')
df['vict_sex'] = df['vict_sex'].map({'M': 'Male', 'F': 'Female'}).fillna('Unknown')
print(df[['vict_descent', 'vict_sex']])
```

```python
df=df.rename(columns={'lat':'latitude','lon':'latitude','crm_cd':'crime_code','rpt_dist_no':'district_no','dr_no':'doc_no','vict_descent':'vict_race'})
print(df.columns)
```

```python
df.to_csv('Cleaned_Crime_Data_from_2010_to_Present.csv', index=False)
     
```
